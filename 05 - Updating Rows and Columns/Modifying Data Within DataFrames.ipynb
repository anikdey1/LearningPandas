```python
def map_manually(input_function, input_iterable):
    result_iterable = []

    for input_item in input_iterable:
        result_item = input_function(input_item)
        result_iterable.append(result_item)

    return result_iterable

number_list = [2, 3, 5, -5, 777]

manual_map_result = map_manually(lambda x: x**3, number_list)
print(manual_map_result)
```

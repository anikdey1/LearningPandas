```python
import pandas as pd
```

```python
people_dict = {
    'id': ['1001', '1002', '1003', '2001', '2002', '2003'],
    'first_name': ['Alice', 'Bartholomew', 'Claire', 'Dave', 'Emma', 'Freya'],
    'last_name': ['Zane', 'Yeoman', 'Xavier', 'Weiss', 'Vichy', 'Underwood'],
    'email_address': ['a_zane@email.com', 'b_yeoman@email.com', 'c_xavier@email.com', 'dave_w@email.com', 'emma_v@email.com', 'freya_u@email.com']
    }

people_df = pd.DataFrame(people_dict)
people_df
```

```python
people_df['first_name'] + ' ' + people_df['last_name']
```

```python
people_df['full_name'] = people_df['first_name'] + ' ' + people_df['last_name']

people_df
```

```python
people_df.drop(columns=['first_name', 'last_name'], inplace=True)
```

```python
people_df
```

```python
people_df['full_name'].str.split(' ')
```

```python
people_df['full_name'].str.split(' ', expand=True)
```

```python
people_df[['first_name', 'last_name']] = people_df['full_name'].str.split(' ', expand=True)

people_df
```

```python
new_row = {
    'id': 3001,
    'email_address': 'g_rhoades@email.com',
    'full_name': 'Gary Rhoades',
    'first_name': 'Gary',
    'last_name': 'Rhoades'
    }

new_row_df = pd.DataFrame.from_records([new_row])
new_row_df
```

```python
people_df = pd.concat([people_df, new_row_df])
people_df
```

```python
new_row_2 = {
    'id': [3002],
    'email_address': ['h_rhoades@email.com'],
    'full_name': ['Harry Rhoades'],
    'first_name': ['Harry'],
    'last_name': ['Rhoades']
    }

new_row_df_2 = pd.DataFrame(new_row_2)
new_row_df_2
```

```python
people_df = pd.concat([people_df, new_row_df_2])
people_df
```

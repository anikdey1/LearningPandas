```python
country_response_counts = sr_df['Country'].value_counts()
country_response_counts
```

```python
country_python_counts = country_group['LanguageHaveWorkedWith'].apply(lambda item: item.str.contains('Python').sum())
country_python_counts
```

```python
country_python_response_counts = pd.concat([country_response_counts, country_python_counts], axis='columns', sort=False)
country_python_response_counts
```

```python
renamed_columns = {
    'count': 'Response Count',
    'LanguageHaveWorkedWith': 'Python Count'
}

country_python_response_counts.rename(columns=renamed_columns, inplace=True)
country_python_response_counts
```

```python
country_python_response_counts['Python Percentage'] = country_python_response_counts['Python Count'] / country_python_response_counts['Response Count'] * 100
country_python_response_counts
```
